# Guidestar Targets Example

Some programs uses multiple guidestar targets for a given observation. We can observe the technical functions of the JWST when FGS locks on to a guidestar and switches to the next using Spelunker. For this example, we will be exploring Program ID 2079.

```python
import sys
sys.path.append('/Users/ddeal/JWST-Treasure-Chest-2023/JWST-FGS-Spelunker-Repos/JWST-FGS-Spelunker/')

import Spelunker

spk = Spelunker.load('/Users/ddeal/JWST-Treasure-Chest-2023/', pid=1534)
```

```python
spk.fg_table['filenames'][0][0:20]
spk.fg_table['filenames'][0][20:42]
```

```python
ra = spk.object_properties['ra']
dec = spk.object_properties['dec']
```

```python
import matplotlib.pyplot as plt

plt.plot(ra, dec)
```

```python
plt.plot(spk.fg_flux)
```

```python
spk.mast_api_token = '8eedd0a371a44e598a9c4489e3148811'

fig, ax = plt.subplots(figsize=(12,4))
ax = spk.mnemonics_local('FILENAME')
ax = spk.mnemonics_local('GUIDESTAR')
plt.xlim(60067.84, 60067.9)
```

```python
import matplotlib.pyplot as plt
```

```python
ax.transAxes
```

```python
import matplotlib.transforms as transforms
#trans = transforms.blended_transform_factory(ax.transData, ax.transAxes) #https://stackoverflow.com/a/63153806

ax = plt.subplot()
trans = transforms.blended_transform_factory(ax.transData, ax.transAxes)

for (x, y) in zip(spk.guidestar_mnemonics[0], spk.guidestar_mnemonics[1]):
    ax.axvline(x, color='lightgreen')
    ax.text(x, 0.9, s=y, transform=trans)
```
